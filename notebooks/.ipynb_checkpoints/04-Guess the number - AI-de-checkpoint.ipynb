### Übung 04, Rückgabe eines Wertes
Bei welchem Funktionsaufruf wird der Wert gespeichert oder verwendet, und bei welchem Aufruf bleibt er ungenutzt und ist somit für immer verloren (verschwindet wahrscheinlich im nächsten schwarzen Loch)?

**Antworts:**

- 1. 
- 2.
- 3. 

```python
def compute_age(birth_year, current_year):
    """
    Berechne das Alter anhand des Geburtsjahres und des aktuellen Jahres.

    Parameters
    ----------
    birth_year : int
    current_year : int
    
    Returns
    ----------
    int
        age
    """
    return current_year - birth_year

# 1.
my_age = compute_age(1976, 2020)
print(my_age)

# 2.
compute_age(1976, 2020)

# 3.
if compute_age(1976, 2020) > 40:
    print("Du bist definitiv über 40!")
else:
    print("Als ich in deinem Alter war...")
```

```python
def do_something(x):
    """
    Tut etwas!
    """
    x = x + 2
    return x
    
def yet_another_function(x):
    """
    Noch einer!
    """
    return x * 2
    
x = yet_another_function(3)
while x < 20:
    x = do_something(x)
    if x == 13:
        break
print(x)
```
